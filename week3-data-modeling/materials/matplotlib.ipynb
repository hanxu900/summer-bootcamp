## Basic Line Plots
Visualization of a single function $y = f(x)$.

### Create figure and axes

```python
%matplotlib inline
plt.style.use('seaborn-whitegrid')
fig = plt.figure()
ax = plt.axes()
```

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 2 * np.pi, 1000)
ax.plot(x, np.sin(x));
```

Alternatively, we can use the pylab interface and let the figure and axes be created for us in the background.

```python
plt.plot(x, np.sin(x));
```

```python
plt.plot(x, np.sin(x));
plt.plot(x, np.cos(x));
```

### Plot with line color and styles

```python
plt.plot(x, np.sin(x - 0), color='blue')        # specify color by name
plt.plot(x, np.sin(x - 1), color='g')           # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 to 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
```

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':');  # dotted
```

### Customize Axes Settings
Matplotlib does a decent job of choosing default axes limits for your plot, but sometimes it's nice to have finer control. 
* ``plt.xlim()``
* ``plt.ylim()``

```python
plt.plot(x, np.sin(x))

plt.xlim(- np.pi, 3 * np.pi)
plt.ylim(-1.5, 1.5);
```

```python
plt.plot(x, np.sin(x))

plt.xlim(3 * np.pi, - np.pi) # reverse the axis
plt.ylim(-1.5, 1.5);
```

Set x and y limit with a single ``plt.axis()`` call

```python
plt.plot(x, np.sin(x))
plt.axis([- np.pi, 3 * np.pi, -1.5, 1.5]);
```

### Labeling Plots

```python
plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)");
```

Create a plot legend that labels each line type through ``plt.legend()`` method. 

```python
plt.plot(x, np.sin(x), '-b', label='sin(x)')
plt.plot(x, np.cos(x), ':g', label='cos(x)')

plt.legend();
```

## Basic Scatter Plots
Instead of points being joined by line segments, here the points are represented individually with a dot, circle, or other shape. 

```python
x = np.linspace(0, 2 * np.pi, 20)
y = np.sin(x)

plt.plot(x, y, 'o', color='black');
```

```python
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker,
             label="marker='{0}'".format(marker))
plt.legend(numpoints=1)
plt.xlim(0, 1.8);
```

### Scatter Plots with ``plt.scatter``
Similar to ``plt.plot`` function

```python
plt.scatter(x, y, marker='o');
```

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,  # alpha for adjust transparency level
            cmap='viridis')  # perceptually uniform shades of blue-green-yellow
plt.colorbar();  # show color scale
```

```python
plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,  # alpha for adjust transparency level
            cmap='magma')  # perceptually uniform shades of black-red-white
plt.colorbar();  # show color scale
```

For more on colormaps, please refer to [Matplotlib Official Doc](https://matplotlib.org/api/pyplot_summary.html?highlight=colormaps#matplotlib.pyplot.colormaps)

### ``plot`` vs ``scatter``
* Efficiency - ``plt.plot`` can be noticeably more efficient than ``plt.scatter`` when dataset size is large
* Reason is that ``plt.scatter`` has the capability to render a different size and/or color for each point, so the renderer must do the extra work of constructing each point individually.

## Example: Stock Analysis

```python
import seaborn as sns
from pandas_datareader import data

stock_list = ['AAPL', 'GOOG', 'AMZN', 'MSFT', 'FB']

stock = data.DataReader(stock_list, start='2000', end='2019',
                       data_source='yahoo')
stock.head()
```

```python
type(stock)
```

```python
type(stock['High'])
```

```python
type(stock['High']['AAPL'])
```

```python
stock['High'].describe()
```

```python
stock['High'].info()
```

```python
stock['High'].plot(figsize=(13, 8))
```

```python
stock['Volume'].plot(figsize=(13, 8))
```

### What's the daily return average of a stock?

```python
stock['Adj Close'].head()
```

```python
stock['Adj Close'].pct_change().tail()
```

```python
stock['Adj Close'].pct_change().plot(alpha = 0.5, figsize = (15, 10))
```

```python
stock['Adj Close'][['AAPL', 'FB']].pct_change().plot(alpha = 0.5, figsize = (15, 10))
```

```python
sns.distplot(stock['Adj Close']['AAPL'].pct_change().dropna(), color='red', bins=100)
```

```python
stock_pct_change = stock['Adj Close'].pct_change()
stock_pct_change.tail()
```

```python
sns.jointplot('GOOG', 'GOOG', stock_pct_change, kind='scatter', color='green')
```

```python
sns.jointplot('GOOG', 'AAPL', stock_pct_change, kind='scatter')
```

```python
sns.jointplot('GOOG', 'AAPL', stock_pct_change, kind='hex')
```

```python
sns.jointplot('GOOG', 'AAPL', stock_pct_change, kind='kde')
```

```python
sns.jointplot('GOOG', 'AAPL', stock_pct_change, kind='reg')
```

```python
sns.pairplot(stock_pct_change)
```
